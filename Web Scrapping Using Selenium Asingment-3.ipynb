# 2. In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product vertical has less than 3 pages in search results then scrape all the products available under that product vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.


```python
# scrape all product urls
product_urls = []
for page in range(0,3):
    #for loop for scrapping 3 page
    url=driver.find_elements_by_xpath('//a[@class="a-link-normal a-text-normal"]')  
    #scraping urls
    for i in url:
        product_urls.append(i.get_attribute("href"))     
        #appending the urls in product_urls list
    next_button=driver.find_element_by_xpath('//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')   
    #scraping the list of buttons from the page
    next_button.click()
    time.sleep(2)

#Make empty lists
Brand = []   
Name = []
Rating = []
number_ratings = []
Price = []
Return = []
expected_delivery = []
Availability = [] 
Other_Details = []
prod_url=[]

for url in product_urls:
    driver.get(url)
    time.sleep(2)
    
    # Extracting Brand from xpath
    try:
        brand=driver.find_element_by_xpath("//td[@class='a-span9']/span")     
        Brand.append(brand.text)
    except NoSuchElementException:
        Brand.append('-')    


    time.sleep(1)
    # Extracting name from id
    try:
        name = driver.find_element_by_id('productTitle')      
        Name.append(name.text)
    except NoSuchElementException:
        Name.append('-')


    time.sleep(1)    
    # Extracting Ratings from xpath    
    try:
        rating = driver.find_element_by_xpath('//span[@class="a-size-base a-nowrap"]//span')  
        Rating.append(rating.text)
    except NoSuchElementException:
        Rating.append('-')


    time.sleep(1)
    # Extracting No of Ratings from xpath
    try:
        no_rating = driver.find_element_by_id('acrCustomerReviewText')  
        number_ratings.append(no_rating.text)
    except NoSuchElementException:
        number_ratings.append('-')


    time.sleep(1)
    # Extracting price from xpath    
    try:
        price = driver.find_element_by_xpath('//td[@class="a-span12"]')  
        Price.append(price.text)
    except NoSuchElementException:
        Price.append('-')
        
    time.sleep(1)
    # Extracting Return from xpath
    try:
        return_ = driver.find_element_by_xpath('//a[@class="a-size-small a-link-normal a-text-normal"]')  
        Return.append(return_.text)
    except NoSuchElementException:
        Return.append('-')


    time.sleep(1)
    # Extracting expected_delivery from xpath
    try:
        expected_d= driver.find_element_by_xpath('//div[@id="ddmDeliveryMessage"]/b')  
        expected_delivery.append(expected_d.text)
    except NoSuchElementException:
         expected_delivery.append('-')


    time.sleep(1)
    # Extracting Availability from xpath         
    try:
        availabal= driver.find_element_by_xpath('//span[@class="a-size-medium a-color-success"]')  
        Availability.append(availabal.text)
    except NoSuchElementException:
        Availability.append('-')

    time.sleep(1)
    # Extracting Other Details from xpath    
    try:
        other_details= driver.find_element_by_xpath('//ul[@class="a-unordered-list a-vertical a-spacing-mini"]')  
        Other_Details.append(other_details.text)
    except NoSuchElementException:
        Other_Details.append('-')

        
    time.sleep(2)  
     # getting the product url
    try:
        product_url=driver.find_element_by_xpath("//div[@class='a-section a-spacing-none icon-content']/a")
        prod_url.append(product_url.get_attribute("href"))
    except NoSuchElementException:
        prod_url.append('-')


# Make dataframe
df = pd.DataFrame({})
df['Brand Name'] = Brand
df['Product Name'] = Name
df['Rating'] = Rating
df['Rating Number'] = number_ratings
df['Price'] = Price
df['Return/Exchange'] = Return
df['Expected Delivery'] = expected_delivery
df['Availability'] = Availability
df['Other Details'] = Other_Details     
df['Product url'] = prod_url
```

# 4. Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”, “Secondary Camera”, “Display Size”, “Display Resolution”, “Processor”, “Processor Cores”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the details is missing then replace it by “- “. Save your results in a dataframe and CSV.

```python
driver=webdriver.Chrome("chromedriver.exe")
# getting the web page with the provided url
url = "https://www.flipkart.com/"
driver.get(url)
time.sleep(3)  # waiting for 3 seconds here
crosslogin_btn = driver.find_element_by_xpath("//div[@class='_2QfC02']//button").click()

#giving inputs to search bar
search_bar = driver.find_element_by_xpath("//div[@class='_3OO5Xc']//input")
search_bar.send_keys("Smartphones")

click_search_btn = driver.find_element_by_xpath("/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/button")
click_search_btn.click()
time.sleep(5)

# fetching urls of every smartphone on Flipkart
urls = []
for i in driver.find_elements_by_xpath("//a[@class='_1fQZEK']"):
    urls.append(i.get_attribute("href"))  
# getting the required data inside the empty lists

Phone_Name = []
Brand_Names=[]
Phone_Color = []
RAM = []
Storage = []
Primary_Camera = []
Secondry_Camera = []
Display_Size = []
Display_Resolution = []
Processor = []
Processor_Cores = []
Battery_Capacity = []
Prices = []
product_url = []  


'''Fetching Phone name'''
for links in urls:
    driver.get(links)
    try:
        phone_name =driver.find_element_by_xpath("//span[@class='B_NuCI']")   
        Phone_Name.append(phone_name.text)
    except NoSuchElementException:
        Phone_Name.append('-')  
        
    '''Fetching Brand name'''
    try:
        brand_name = driver.find_element_by_xpath("/html/body/div/div/div[3]/div[1]/div[2]/div[1]/div[1]/div/div[4]/a")
        Brand_Names.append(brand_name.text.replace('Mobiles',''))
    except NoSuchElementException:
        Brand_Names.append('-')
        
   
    '''Fetching phone color'''
    try:
        color = driver.find_element_by_xpath("//table[@class='_14cfVK']//tr[4]//td[2]")
        Phone_Color.append(color.text)
    except NoSuchElementException:
        Phone_Color.append('-')


        
    '''Fetching RAM''' 
   
    try:
        ram = driver.find_element_by_xpath("//div[@class='_2cM9lP']//div[2]//ul//li[1]")
        RAM.append(ram.text[0:5])
    except NoSuchElementException:
        RAM.append('-')


    '''Fetching Storage'''

    try:
        rom = driver.find_element_by_xpath("//div[@class='_2cM9lP']//div[2]//ul//li[1]")
        Storage.append(rom.text[11:17])
    except NoSuchElementException:
        Storage.append('-')

        
    '''Fetching primary camera'''
    
    try:
        front_camera= driver.find_element_by_xpath("//div[@class='_2cM9lP']//div[2]//ul//li[3]")
        Primary_Camera.append(front_camera.text[0:(front_camera.text).find('|')])
    except NoSuchElementException:
        Primary_Camera.append('-')
    

    '''Fetching Secondry_Camera'''
    try:
        sec_camera= driver.find_element_by_xpath("//div[@class='_2cM9lP']//div[2]//ul//li[3]")
        Secondry_Camera.append(sec_camera.text[(sec_camera.text).find('|')+1 :])
    except NoSuchElementException:
        Secondry_Camera.append('-')

        
    '''Clicking on read more'''

    try:
        read_more= driver.find_element_by_xpath('//button[@class="_2KpZ6l _1FH0tX"]')
        read_more.click()
    except NoSuchElementException:
        pass
    
    '''Fetching Display size'''
    try:
        display= driver.find_element_by_xpath('//div[@class="_1UhVsV"]//div[2]//table//tr[1]//td[2]//li')
        Display_Size.append(display.text)
    except NoSuchElementException:
        Display_Size.append('-')
        
        
        
    '''Fetching Display_Resolution'''
    try:
        display_res= driver.find_element_by_xpath('//div[@class="_1UhVsV"]//div[2]//table//tr[2]//td[2]//li')      
        Display_Resolution.append(display_res.text)
    except NoSuchElementException:
        Display_Resolution.append('-')
    
    
    '''Fetching Processor'''
    try:
        processor= driver.find_element_by_xpath('//div[@class="_1UhVsV"]//div[3]//table//tr[2]//td[2]//li') 
        Processor.append(processor.text)
    except NoSuchElementException:
        Processor.append('-')

 
    '''Fetching Processor cores'''
    try:
        processor_core= driver.find_element_by_xpath('//div[@class="_1UhVsV"]//div[3]//table//tr[3]//td[2]//li') 
        Processor_Cores.append(processor_core.text)
    except NoSuchElementException:
        Processor_Cores.append('-')

    '''Fetching battery_capacity'''
    try:
        battery_capacity = driver.find_element_by_xpath('//div[@class="_1UhVsV"]//div[8]//table//tr[1]//td[2]//li') 
        Battery_Capacity.append(battery_capacity.text)
    except NoSuchElementException:
        Battery_Capacity.append('-')

    '''Fetching Prices'''
    try:
        price = driver.find_element_by_xpath('//div[@class="_30jeq3 _16Jk6d"]')
        Prices.append(price.text)
    except NoSuchElementException:
        Prices.append('-')

df = pd.DataFrame({})
df['Phone Name']=Phone_Name
df['Brand'] = Brand_Names
df['Color'] = Phone_Color
df['RAM'] = RAM
df['Storage']=Storage
df['Primary Camera'] = Primary_Camera
df['Secondry Camera'] = Secondry_Camera
df['Display Size'] = Display_Size
df['Display resolution']=Display_Resolution
df['Processor'] = Processor
df['Cores']=Processor_Cores
df['Battery Capacity']=Battery_Capacity
df['Price']=Prices
df['Url']=urls
df
```

# 5. Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps.

```python
# Activating the chrome browser
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(2)

# opening google maps
url = "https://www.google.co.in/maps"
driver.get(url)
time.sleep(2)

Search = driver.find_element_by_id("searchboxinput")
# locating search bar
Search.clear() 
# clearing search bar
time.sleep(2)
# Asking input from the user
City_input =input("Enter the city name which you want to search on Google Maps:")
Search.send_keys(City_input)  
# entering values in search bar
button = driver.find_element_by_id("searchbox-searchbutton") 
# locating search button
button.click()     
# clicking search button
time.sleep(3)

# storing current page url in new variable
current_page_url = driver.current_url
try:
    lat = re.findall('(\d+(\.\d*)?)', current_page_url)[0]   # fetching the latitude data
    lon = re.findall('(\d+(\.\d*)?)', current_page_url)[1]  # fetching the longitude data
    print("Geospatial Coordinates for", City_input + " on Google Maps are: ")
    print("Latitude  :", lat[0])
    print("Longitude :", lon[0])
    
except Exception as e:
    print("Error: ", str(e))
```

# 7. Write a program to scrap all the available details of best gaming laptops from digit.in.

```python
driver = webdriver.Chrome("chromedriver.exe")
# getting the web page with given url
url = "https://www.digit.in/"
driver.get(url)
time.sleep(3)

# now we will search for best gaming laptops
gaming_laptop = driver.find_element_by_xpath("//div[@class='listing_container']//ul//li[9]")
gaming_laptop.click()
time.sleep(4)

# creating empty lists to store all gaming laptop information
lap_name = []
op_stm = []
display = []
processor = []
memory = []
weight = []
dimensions = []
graph_processor = []
prices = []

# fetching the laptop names
name = driver.find_elements_by_xpath("//table[@id='summtable']//tr//td[1]")
for i in name:
    lap_name.append(i.text)
    
# fetching information related to operating system
try:
    os = driver.find_elements_by_xpath("//div[@class='Spcs-details']//tr[3]//td[3]")
    for i in os:
        op_stm.append(i.text)
except NoSuchElementException:
    op_stm.append('-')

# fetching information for display
try:
    disp = driver.find_elements_by_xpath("//div[@class='Spcs-details']//tr[4]//td[3]")
    for i in disp:
        display.append(i.text)
except NoSuchElementException:
    display.append('-')

# fetching information for processor
try:
    pro = driver.find_elements_by_xpath("//div[@class='Spcs-details']//tr[5]//td[3]")
    for i in pro:
        processor.append(i.text)
except NoSuchElementException:
    processor.append('-')

# fetching information for memory
try:
    mem = driver.find_elements_by_xpath("//div[@class='Spcs-details']//tr[6]//td[3]")
    for i in mem:
        memory.append(i.text)
except NoSuchElementException:
    memory.append('-')

# fetching information for weight
try:
    wgt = driver.find_elements_by_xpath("//div[@class='Spcs-details']//tr[7]//td[3]")
    for i in wgt:
        weight.append(i.text)
except NoSuchElementException:
    weight.append('-')

# fetching information for dimensions
try:
    dim = driver.find_elements_by_xpath("//div[@class='Spcs-details']//tr[8]//td[3]")
    for i in dim:
        dimensions.append(i.text)
except NoSuchElementException:
    dimensions.append('-')

# fetching information for graphic processor
try:
    g_pro = driver.find_elements_by_xpath("//div[@class='Spcs-details']//tr[9]//td[3]")
    for i in g_pro:
        graph_processor.append(i.text)
except NoSuchElementException:
    graph_processor.append('-')

# fetching information for price
try:
    price = driver.find_elements_by_xpath("//td[@class='smprice']")
    for i in price:
        prices.append(i.text.replace('₹','Rs '))
except NoSuchElementException:
    prices.append('-')
    
    
# Creating dataframe    
df=pd.DataFrame({})    
df['laptop_name']=lap_name
df['operating_system']=op_stm
df['Display']=display
df['Processor']=processor
df['Memory']=memory
df['Weight']=weight
df['Dimensions']=dimensions
df['Graph_processor']=graph_processor
df['Prices']=prices
df
```

# 9. Write a program to extract at least 500 Comments, Comment upvote and time when comment was posted from any YouTube Video.

```python
driver = webdriver.Chrome('chromedriver.exe')
# obtaining the web page with given url
url = "https://www.youtube.com/"
driver.get(url)

# obtaining the search option on our web driver
search_bar = driver.find_element_by_xpath('//input[@id="search"]')

# obtaining the item to be searched from the user
user_input = str(input("Enter what you want to search on YouTube: "))
search_bar = driver.find_element_by_xpath ('//input[@id="search"]')
search_bar.send_keys(user_input)

# performing a click using xpath function in our web driver
search_btn = driver.find_element_by_xpath('//*[@id="search-icon-legacy"]')
search_btn.click()
time.sleep(3)

# now we will click on the YT video
video = driver.find_element_by_xpath("//a[@id='video-title']").click()
time.sleep(10)

```

```python
# we are going to scroll through the comments
for _ in range(1000):
    driver.execute_script("window.scrollBy(0,1000)")
# fetching all the comments
comments = []
try:
    comment = driver.find_elements_by_xpath("//*[@id='content-text']")[:500]
    for i in comment:
        comments.append(i.text)
except NoSuchElementException:
    comments.append('-')
time.sleep(5)

# fetching all the upvotes
upvotes = []
try:
    up = driver.find_elements_by_xpath("//*[@id='vote-count-middle']")[:500]
    for i in up:
        upvotes.append(i.text)
except NoSuchElementException:
    upvotes.append('-')
time.sleep(5)

# fetching all the time information indicating when the comment was made
times = []
try:
    timing = driver.find_elements_by_xpath("//yt-formatted-string[@class='published-time-text above-comment style-scope ytd-comment-renderer']")[:500]
    for i in timing:
        times.append(i.text.replace('(edited)',''))
except NoSuchElementException:
    times.append('-')
time.sleep(5)

# creating dataframe
df=pd.DataFrame({})
df['Comments']=comments
df['upvotes']=upvotes
df['Time']=times
```

# 10. Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in “London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall reviews, privates from price, dorms from price, facilities and property description.

```python
driver = webdriver.Chrome('chromedriver.exe')
# obtaining the web page from the given url
url = "https://www.hostelworld.com/"
driver.get(url)
```

```python
# cross_ad = driver.find_element_by_xpath('/html/body/div[6]/div[1]')
# cross_ad.click()
```

```python
# now we will find the search bar and search for hostels in London location
search_bar = driver.find_element_by_xpath('//*[@id="search-input-field"]')
search_bar.send_keys('London')
```

```python
click_london_button = driver.find_element_by_xpath('//*[@id="predicted-search-results"]/li[2]')
click_london_button.click()
```

```python
lets_go_button = driver.find_element_by_xpath('//*[@id="search-button"]')
lets_go_button.click()
```

```python
# creating empty lists to collect the required information
hostel_name = []
distance = []
pvt_prices = []
dorms_price = []
rating = []
reviews = []
over_all = []
facilities = []
description =[]
product_url = []

for i in driver.find_elements_by_xpath("//div[@class = 'pagination-item pagination-current' or @class='pagination-item']"):
    i.click()
    time.sleep(4)
    
    # fetching hostel name
    try:
        name = driver.find_elements_by_xpath("//h2[@class='title title-6']")
        for i in name:
            hostel_name.append(i.text)
    except NoSuchElementException:
        hostel_name.append('-')
        
    # fetching the distance from city centre    
    try:
        dist = driver.find_elements_by_xpath("//div[@class='subtitle body-3']//a//span[1]")
        for i in dist:
            distance.append(i.text.replace('Hostel - ',''))
    except NoSuchElementException:
        distance.append('-')
        
    for i in driver.find_elements_by_xpath("//div[@class='prices-col']"):
        
    # fetching the privates from price details
        try:
            pvt_price = driver.find_element_by_xpath("//a[@class='prices']//div[1]//div")
            pvt_prices.append(pvt_price.text)
        except NoSuchElementException:
            pvt_prices.append('-')
            
    # fetching the dorms from price information
    for i in driver.find_elements_by_xpath("//div[@class='prices-col']"):
        try:
            dorms = driver.find_element_by_xpath("//a[@class='prices']//div[2]//div")
            dorms_price.append(dorms.text)
        except NoSuchElementException:
            dorms_price.append('-')
            
    # fetching all the facilities
    try:
        fac1 = driver.find_elements_by_xpath("//div[@class='has-wifi']")
        fac2 = driver.find_elements_by_xpath("//div[@class='has-sanitation']")
        for i in fac1:
            for j in fac2:
                facilities.append(i.text +', '+ j.text )
    except NoSuchElementException:
        facilities.append('-')
        
    # fetching the urls of each hostel
    p_url = driver.find_elements_by_xpath("//div[@class='prices-col']//a[2]")
    for i in p_url:
        product_url.append(i.get_attribute('href'))

for i in product_url:
    driver.get(i)
    time.sleep(3)
    
    # clicking on the show more button for description
    try:
        driver.find_element_by_xpath("//a[@class='toggle-content']").click()
        time.sleep(5)
    except NoSuchElementException:
        pass
    
    # fetching the ratings for the hostels
    try:
        rat = driver.find_element_by_xpath("//div[@class='score orange big' or @class='score gray big']")
        rating.append(rat.text)
    except NoSuchElementException:
        rating.append('-')
        
    # fetching the total reviews        
    try:
        rws = driver.find_element_by_xpath("//div[@class='reviews']")
        reviews.append(rws.text.replace('Total Reviews',''))
    except NoSuchElementException:
        reviews.append('-')
        
    # fetching the overall review
    try:
        overall_rw = driver.find_element_by_xpath("//div[@class='keyword']//span")
        over_all.append(overall_rw.text)
    except NoSuchElementException:
        over_all.append('-')
        
    # fetching the property description 
    try:
        disc = driver.find_element_by_xpath("//div[@class='content']")
        description.append(disc.text)
    except NoSuchElementException:
        over_all.append('-')
```

```python
# Creating dataframe
df = pd.DataFrame({})
df['hostel_name']=hostel_name
df['distance']=distance
df['pvt_prices']=pvt_prices
df['dorms_price']=dorms_price
df['rating']=rating
df['over_all']=over_all
df['facilities']=facilities
df['description']=description
df['product_url']=product_url
df
```

```python
# Getting url
def find_headers(url):
    response = requests.get(url)
#     Checking Response Status
    if response.status_code != 200:
        raise Exception('Failed to load Page {}'.format(url))
    else:
        page_url = response.text
        #BeautifulSoup 
        page = BeautifulSoup(page_url,'html.parser')       
        # Making Dictionary
        header_tags = [] # empty list
        for header in page.find_all(["h1","h2","h3","h4","h5","h6"]):
            header_tags.append(header.text.replace('\n',''))
        
        df=pd.DataFrame({})
        df['Wikipedia Headers']=header_tags
        return df
   
```

```python
url = 'https://en.wikipedia.org/wiki/Main_Page'
find_headers(url)
```

### 2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release)

```python
# Getting Url
def top_movies(url2):
    response = requests.get(url2)    
#Checking response
    if response.status_code != 200:
        raise Exception('Failed to load Page {}'.format(url2))
    else:         
#BeautifulSoup     
        page_url = response.text
        page = BeautifulSoup(page_url,'html.parser')
# Making Dictionary
        top_movie_dict = { 
            "movies":[],
            'rating':[],
            'year':[]
        }
        for i in range(100):
            movie_title_tag = page.find_all('td',{'class':'titleColumn'})
            rating_tag = page.find_all('td',{'class':'ratingColumn imdbRating'})
            year_tag= page.find_all('span',{'class':'secondaryInfo'})

            top_movie_dict['movies'].append(movie_title_tag[i].contents[1].text)
            top_movie_dict['rating'].append(rating_tag[i].text.strip())
            top_movie_dict['year'].append(year_tag[i].text[1:-1])
            
            
        df = pd.DataFrame(top_movie_dict)
        return df
```

```python
# Function call

url2 = 'https://www.imdb.com/chart/top/?ref_=nv_mv_250'
top_movies(url2)
```

### 4. Write a python program to scrap book name, author name, genre and book review of any 5 books from ‘www.bookpage.com’

```python
def Books(url4):
    response = requests.get(url4)
    # Checking Response Status
    if response.status_code != 200:
        raise Exception('Failed to load Page {}'.format(url4))
    else:    
        page_url = response.text
        #BeautifulSoup 
        page = BeautifulSoup(page_url,'html.parser') 
    return page  

def Top_books(page):
        # Making Dictionary
        book_dict = { 
            "Book_Name":[],
            'Author':[],
            'gener':[],
            'Review':[]
        }
        div = page.find_all('div',{'class':'flex-article-content'})
        for i in range(0,5):
            book_dict['Book_Name'].append(div[i].contents[1].text.strip())
            book_dict['Author'].append(div[i].contents[3].text.strip()) 
            book_dict['Review'].append(div[i].contents[7].text.strip())
            lis=[]
            for j in range(0,5):
                a=div[j].contents[5].find_all('a')
                s=''
                for i in range(len(a)):
                    s= s +  div[j].contents[5].find_all('a')[i].text + ' / '
                lis.append(s)
        for i in lis:
            book_dict['gener'].append(i)            
        df = pd.DataFrame(book_dict)
        return df
```

```python
# Function call

url4 = 'https://bookpage.com/reviews'
Top_books(Books(url4))
```

### 7. Write a python program to scrape details of all the mobile phones under Rs. 20,000 listed on Amazon.in. The scraped data should include Product Name, Price, Image URL and Average Rating.

```python
def Amazone(url15):
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    response = requests.get(url15,headers=headers)
    # Checking Response Status
    if response.status_code != 200:
        raise Exception('Failed to load Page {}'.format(url15))
    else:    
        page_url = response.text
        #BeautifulSoup 
        page = BeautifulSoup(page_url,'html.parser')    
        return page  

def Phones(page):    
    main_container = page.find_all('div',class_='s-main-slot s-result-list s-search-results sg-row')
    # Making Dictionary
    phone_dict={
        'Product_name':[],
        'Price':[],
        'Ratings':[],
        'image_url':[] 
    }
    for items in range(1,15):
        phone_dict['Product_name'].append(main_container[0].find_all('span',class_='a-size-medium a-color-base a-text-normal')[items].get_text())
        phone_dict['Price'].append(main_container[0].find_all('span',class_='a-price-whole')[items].get_text())
        phone_dict['Ratings'].append((main_container[0].find_all('span',class_='a-icon-alt')[items].get_text())[:4])    
        phone_dict['image_url'].append(main_container[0].find_all("img",class_="s-image")[items].get("src"))
        #Creating Dataframe
        df = pd.DataFrame(phone_dict)
    return df       
```

```python
# Function call

url15 = 'https://www.amazon.in/s?k=smartphones+under+20000+rupees+only&crid=3QV1RABX4EPDF&sprefix=smartphones+under+20000+rupees%2Caps%2C369&ref=nb_sb_ss_ts-doa-p_3_30'
Phones(Amazone(url15))
```

### 8. Write a python program to extract information about the local weather from the National Weather Service website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day extended forecast display for the city. The data should include period, short description, temperature and description

```python
# Getting Url
def Forcast(url12):
    response = requests.get(url12)
    # Checking Response Status
    if response.status_code != 200:
        raise Exception('Failed to load Page {}'.format(url12))
    else:    
        page_url = response.text
        #BeautifulSoup 
        page = BeautifulSoup(page_url,'html.parser')    
        return page  
        
def Details(page):
    list_box = page.find_all('li',class_='forecast-tombstone')
    desc = page.find_all('div',class_='col-sm-10 forecast-text')
    # Making Dictionary
    details_dict = {
        "Period":[],
        "short description":[],
        'temperature':[],
        "description":[]
    } 
    for i in range(0,len(list_box)):  
        details_dict["Period"].append(list_box[i].find_all('p')[0].text)
        details_dict["short description"].append(list_box[i].find_all('p')[2].text)
        details_dict["temperature"].append(list_box[i].find_all('p')[3].text)
        details_dict["description"].append(desc[i].text)
        
    #Creating Dataframe
    df=pd.DataFrame(details_dict)
    return df
```

```python
# Function call

url12 = 'https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996#.YR5oq4gzbIV'
Details(Forcast(url12))
```

### 9. Write a python program to scrape fresher job listings from ‘https://internshala.com/’. It should include job title, company name, CTC, and apply date.

```python
# Getting Url
def Internshala():
    #Creating Empty Dataframe
    final=pd.DataFrame()
    for j in range(1,5):
        response = requests.get(f'https://internshala.com/fresher-jobs/page-{j}')
        # Checking Response Status
        if response.status_code != 200:
            raise Exception('Failed to load Page {}'.format(f'https://internshala.com/fresher-jobs/page-{j}'))
        else:    
            page_url = response.text
            #BeautifulSoup 
            page = BeautifulSoup(page_url,'html.parser')    
            Box_content = page.find_all('div',class_='internship_meta')
            # Making Dictionary
            details_dict = {
                "Job title":[],
                "Company name":[],
                'CTC':[],
                "Apply Date":[]
                }
            for i in range(0,len(Box_content)):  
                details_dict["Job title"].append(Box_content[i].find_all('a')[0].text)
                details_dict["Company name"].append(Box_content[i].find_all('a')[1].text.strip())
                details_dict["CTC"].append(Box_content[i].find_all('div',class_="item_body")[1].text.strip())
                details_dict["Apply Date"].append(Box_content[i].find_all('div',class_="item_body")[2].text)
                #Creating Dataframe
                df=pd.DataFrame(details_dict)
            #Append Dataframe      
            final=final.append(df,ignore_index=True)
    return final
```

```python
# Function call

Internshala()
```
